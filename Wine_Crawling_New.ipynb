```python
import requests
from requests import get
import time
import random

from bs4 import BeautifulSoup
import pandas as pd

from IPython.display import clear_output

import math
```

```python
#get beautiful soup from url
def get_soup(url):
    response = requests.get(url)
    if not response.ok: # .status_code == 200:
        print(f'Code:{response.status_code},url:{url}')
        return None
        
    soup = BeautifulSoup(response.text,'html.parser')
    return soup
```

```python
#get list of urls from page of wines
def get_wine_urls_from_page(soup):
    wine_links_list=[]
    wine_soup_list=soup.find_all("div", class_="prodItemInfo")
    for item in wine_soup_list:
        critical_ratings = item.find("span", class_="wineRatings_rating")
        if critical_ratings != None:
            wine_links_list.append(item.find(href=True)['href'])
    #for elem in soup.find_all('a',class_="prodItemInfo_link event_productClick"):
    #    wine_links_list.append(elem['href'])
    
    return wine_links_list
    #return wine_links_list
```

```python
wine_links_list=[]
```

```python
#variables for get links:
start_page = 0
```

```python
#===collect links from number of pages of specific kind of wine===

#white links:
#url_part1='https://www.wine.com/list/wine/white-wine/7155-125/'
#url_part2='?iid=Homepage:Pill:White%20Wine'
#sparkling links:
#url_part1='https://www.wine.com/list/wine/champagne-and-sparkling/7155-123/'
#url_part2='?iid=Homepage:Pill:Sparkling'
#Rose links:
url_part1='https://www.wine.com/list/wine/rose-wine/7155-126/'
url_part2='?iid=Homepage:Pill:Rose'

page=start_page


while len(wine_links_list) <=4000 and page<=255:   #number of pages to search
    url=url_part1+str(page)+url_part2
    wines_page_soup=get_soup(url)
    if wines_page_soup!=None:
        wine_links_list+=get_wine_urls_from_page(wines_page_soup)
        #random delay:
        value = random.random()
        scaled_value = 1+(value*(9-5))
        clear_output(wait=False)
        print(f"{page} : {scaled_value},link counter:{len(wine_links_list)}")
    else:
        print(f"Empty soup.\nLast page:{page}")
        break
    page+=1
#===collect links from number of pages of specific kind of wine===
```

```python
#load Dataframe:
df = pd.read_csv("New_wine_file.csv")
df
```

```python
#variables for get wines:
start_row = len(df)+1
#start_row = 0
last_row = len(df)+len(wine_links_list)
kind = 'rose'
```

```python
#=======================crawling wines specifics(dataframe)==================
#base url:url='https://www.wine.com'
file_name='New_wine_file.csv'
row=start_row

files_count=0 #counter for file saves

while row <last_row:

    url='https://www.wine.com'+wine_links_list[row-start_row]
    wine_soup=get_soup(url)
    
    
    df.loc[df.shape[0]]=[None, None, None,None, None, None,None, None, None,None,None,None,None,None]
    
    value = random.random()
    scaled_value = 1+(value*(9-5))
    clear_output(wait=True)
    print(f"{row} : {scaled_value}")
    #================================
    df.at[row, 'Name'] = str(wine_soup.h1.string)
    df.at[row, 'Price'] = float(wine_soup.find("meta",itemprop="price").get("content"))
    df.at[row, 'Grape Sort'] = wine_soup.find("span",class_="prodItemInfo_varietal").text
    
    Place=wine_soup.find("span",class_="prodItemInfo_originText").find("a",class_="pipOrigin_link")
    if Place == None:
        df.at[row, 'Place'] = None
    else:
        df.at[row, 'Place'] = Place.text
    
    Avg_Rating = wine_soup.find("span",class_="averageRating_average")
    if Avg_Rating == None:
        df.at[row, 'Avg Rating'] = None
    else:
        df.at[row, 'Avg Rating'] = float(Avg_Rating.text)
    
    Ratings = wine_soup.find("span",class_="averageRating_number")
    if Ratings == None:
        df.at[row, 'Ratings'] = None
    else:
        df.at[row, 'Ratings'] = int(Ratings.text)
        
    Product_Volume = wine_soup.find("span",class_="prodAlcoholVolume_text")
    if Product_Volume == None:
        df.at[row, 'Product Volume'] = None
    else:
        df.at[row, 'Product Volume'] = float(Product_Volume.text)
        
    Alcohol = wine_soup.find("span",class_="prodAlcoholPercent_percent")
    if Alcohol == None:
        df.at[row, 'Alcohol %'] = None
    else:
        df.at[row, 'Alcohol %'] = float(Alcohol.text)
    
    df.at[row, 'Wine link'] = url
    
    Winemaker_Notes = wine_soup.find("h2",string="Winemaker Notes").parent.find("div",class_="viewMoreModule_text")
    if Winemaker_Notes == None:
        df.at[row, 'Winemaker Notes'] = None
    else:
        df.at[row, 'Winemaker Notes'] = Winemaker_Notes.text
    
    #collect critical acclaims:
    critical_acclaims=wine_soup.find_all("div",class_="pipSecContent_copy")
    crit_rewiews_on_str=''
    for review in critical_acclaims:
        Acclaim=review.text
        crit_rewiews_on_str+=Acclaim
        
    df.at[row, 'Critical Acclaim'] = crit_rewiews_on_str
    crit_rewiews_on_str=''
    df.at[row, 'Num of Critics'] = len(critical_acclaims)
    df.at[row, 'Wine Kind'] = kind
    
    # get Avg. Critics rating:
    prof_critics = wine_soup.find("ul",class_='wineRatings_list').find_all("span",class_="wineRatings_rating")
    sum_of_ratings=0.0
    for item in prof_critics:
        sum_of_ratings += float(item.text)
        df['Avg. Critics rating'][row] =  sum_of_ratings/len(prof_critics)
        
    #================================
    #Save Dataframe:
    files_count+=1
    if files_count>=100:
        df.to_csv(file_name,index=False)
        files_count=0
        value = random.random()
        scaled_value = 1+(value*(9-5))
    
    row+=1
```
